```python
import PyPDF2
import os
import nltk
nltk.download('punkt')

def convert_pdfs_to_text(path_to_pdfs):
    """
    Convert all PDF files in a directory to a single text string.

    Args:
        path_to_pdfs (str): Path to the directory containing PDF files.

    Returns:
        str: Combined text content from all successfully processed PDF files.

    Raises:
        Exception: Prints error message if a PDF file cannot be processed.
    """
    
    def extract_text(page):
        """
        Extract text from a single PDF page.

        Args:
            page (PyPDF2.PageObject): A page object from PyPDF2.

        Returns:
            str: Extracted text content from the page.
        """
        text = page.extract_text()
        return text

    combined_text = ""
    for filename in os.listdir(path_to_pdfs):
        if filename.endswith(".pdf"):
            file_path = os.path.join(path_to_pdfs, filename)
            try:
                reader = PyPDF2.PdfReader(file_path)
                text = [extract_text(page) for page in reader.pages]
                combined_text += " ".join(text)
            except Exception as e:
                print(f"Failed to process {filename}: {e}")
    return combined_text

```

```python
def count_tokens(text):
    tokens = nltk.word_tokenize(text)
    return len(tokens)
```

```python
path_to_pdfs = "your/path/to/pdfs"
text = convert_pdfs_to_text(path_to_pdfs)

print(text)
print(f"ntokens: {count_tokens(text)}")
```
